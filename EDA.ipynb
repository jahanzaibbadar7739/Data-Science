```python
# visualize and explore
sns.countplot(data=df, x="survived", hue="age")
```

```python
sns.countplot(data=df, x="survived", hue="sex")
```

```python
sns.barplot(data=df,x="survived", y="fare", hue="who")
```

```python
sns.countplot(x="pclass",data=df)
```

```python
#data distribution
sns.histplot(data=df, x="age", bins=5)
```

```python
sns.histplot(data=df, x="age", bins=5)
```

```python
sns.boxplot(data=df, y="age", x="survived", hue="who")
```

```python
sns.boxplot(data=df, y="age", x="survived", hue="sex")
```

```python
#scatterplot
sns.scatterplot(data=df, x="age", y="fare", hue="pclass")
plt.ylim(0,350)
```

```python
sns.pairplot(df, hue="survived")
```

```python
# stacked barchart/plot
df.groupby(['pclass', 'who'])['fare'].mean().unstack().plot(kind='bar', stacked=True)
```

```python
# swarm plot
sns.swarmplot(data= df, x='survived', y='fare')
```

```python
sns.swarmplot(data= df, x='survived', y='age', hue="who")
```

```python
sns.catplot(data=df,x='pclass', y='fare', hue ='embarked', kind= 'point')
```

```python
#relation of two numeric variables
sns.lmplot(data=df,y='fare', x= 'age', hue='who')
#assignment how to take data for fare less than 300 to remove outlier
```
